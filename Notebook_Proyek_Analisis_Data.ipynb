## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

### Cleaning Data

#### Fixing ... problem

```python
df_hour['dteday'] = pd.to_datetime(df_hour['dteday'])
df_day['dteday']  = pd.to_datetime(df_day['dteday'])

season_map     = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
weather_map    = {1: 'Clear', 2: 'Cloudy/Mist', 3: 'Light Rain/Snow', 4: 'Heavy Rain/Storm'}
yr_map         = {0: '2011', 1: '2012'}
holiday_map    = {0: 'Non-Holiday', 1: 'Holiday'}
workingday_map = {0: 'Non-Workday', 1: 'Workday'}

for df in [df_hour, df_day]:
    df['season_label']     = df['season'].map(season_map)
    df['weather_label']    = df['weathersit'].map(weather_map)
    df['year_label']       = df['yr'].map(yr_map)
    df['holiday_label']    = df['holiday'].map(holiday_map)
    df['workingday_label'] = df['workingday'].map(workingday_map)

def categorize_hour(h):
    if 5 <= h <= 9:
        return 'Pagi (05-09)'
    elif 10 <= h <= 14:
        return 'Siang (10-14)'
    elif 15 <= h <= 19:
        return 'Sore (15-19)'
    elif 20 <= h <= 23 or 0 <= h <= 4:
        return 'Malam (20-04)'
    else:
        raise ValueError(f'Jam tidak valid: {h}')

df_hour['time_category'] = df_hour['hr'].apply(categorize_hour)

print('Data cleaning selesai!')
print(df_hour[['dteday','season_label','weather_label','year_label','time_category','cnt']].head())
```

```python
df_day['year_month'] = df_day['dteday'].dt.to_period('M')
monthly = df_day.groupby('year_month')['cnt'].sum().reset_index()
monthly.columns = ['year_month', 'total_cnt']
monthly['mom_growth'] = monthly['total_cnt'].pct_change() * 100
monthly['year_month_str'] = monthly['year_month'].astype(str)

print('Data bulanan siap!')
monthly.head()
```

**Insight:**
- Kolom `dteday` berhasil dikonversi ke tipe datetime pada kedua dataset,
  memungkinkan analisis berbasis waktu seperti tren bulanan dan MoM growth.
- Mapping label kategorikal (`season_label`, `weather_label`, `year_label`,
  `holiday_label`, `workingday_label`) berhasil diterapkan pada kedua dataset,
  data kini lebih mudah dibaca dan divisualisasikan.
- Kolom `time_category` berhasil ditambahkan pada dataset jam dengan segmentasi
  Pagi, Siang, Sore, dan Malam — terlihat dari output head() yang menampilkan
  'Malam (20-04)' untuk jam 0–4 dengan benar.
- Agregasi bulanan berhasil dibuat — bulan pertama (2011-01) memiliki total
  38.189 peminjaman, lalu naik signifikan ke 48.215 (Feb) dan terus tumbuh
  hingga 135.821 (Mei 2011).
- Kolom `mom_growth` berhasil dihitung — pertumbuhan MoM pertama yang tercatat
  adalah 26.25% (Feb 2011), diikuti 32.83% (Mar) dan 48.13% (Apr),
  menunjukkan pertumbuhan pesat di awal tahun 2011.
- Nilai NaN pada baris pertama `mom_growth` adalah wajar karena tidak ada
  bulan sebelumnya sebagai pembanding.

```python
weather_avg = df_day.groupby('weather_label')['cnt'].mean().reset_index()
weather_avg.columns = ['Kondisi Cuaca', 'Rata-rata Peminjaman']
weather_avg = weather_avg.sort_values('Rata-rata Peminjaman', ascending=False)

baseline = weather_avg[weather_avg['Kondisi Cuaca']=='Clear']['Rata-rata Peminjaman'].values[0]
weather_avg['Penurunan (%)'] = ((weather_avg['Rata-rata Peminjaman'] - baseline) / baseline * 100).round(2)
print(weather_avg)
```
